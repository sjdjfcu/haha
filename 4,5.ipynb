```r
x<-outer(1:9,1:9)
y<-outer(1:9,x)
z<-outer(y,y,"-")
z
```

```r
f<-table(z)
f/9^6
```

```r
plot(as.numeric(names(f)),f,type="h",xlab="行列式值",ylab="频数")
plot(as.numeric(names(f)),f/9^6,type='h',xlab='行列式值',ylab='频率')
```

```r
x<-seq(-3,3,length=20)
y<-seq(-3,3,length=20)
f=function(x,y){(1/2*pi)*exp(-0.5*(x^2+y^2))}
z<-outer(x,y,f)
persp(x,y,z)
```
